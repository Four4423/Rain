```python
import requests_cache
from retry_requests import retry
import openmeteo_requests
from openmeteo_sdk.Variable import Variable
from datetime import datetime
import requests
```

**В качестве датасет выступают данные, взятые с сайта Kaggle**

```python
df = pd.read_csv('/content/Superstore Sales Dataset.csv')
```

**В датасете присутствует столбце 'City', с помощью  APi openweathermap, получим его координаты (функция get_coord_from_city_name)**

```python
cache_session = requests_cache.CachedSession('.cache', expire_after = -1)
retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
openmeteo = openmeteo_requests.Client(session = retry_session)
```

```python
def get_coord_from_city_name(city):
  api_key = ''
  base_url = "http://api.openweathermap.org/data/2.5/weather?"

  complete_url = base_url + "appid=" + api_key + "&q=" + city

  response = requests.get(complete_url)

  if response.status_code == 200:
    data = response.json()
    coord = [data.get('coord', {}).get('lat'), data.get('coord', {}).get('lon')]
    return coord
  else:
    return 'error'
```

**Пример работы функции get_coord_from_city_name:**

```python
get_coord_from_city_name('Moscow')
```

**Теперь применим функцию ко всему датасету и сохраним координаты в Series**

```python
lat_df = df['City'].apply(get_coord_from_city_name)
lat_df
```

**Добавим к датафрейму столбцы 'Latitude' и 'Longitude', значения берем из ранее созданной Series**

```python
df['Latitude'] = lat_df.apply(lambda x: x[0])
df['Longitude'] = lat_df.apply(lambda x: x[1])
```

**После этого, датафрейм вылядит следующим образом:**

```python
df
```

**Добавим функцию, которая будет принимать дату и координаты и возвращать интересующую информацию о погоде. Обращение идет к API open-meteo**

```python
def get_weather_with_coord1(date, lat, lon):
  url = "https://archive-api.open-meteo.com/v1/archive"
  params = {
  "latitude": lat,
  "longitude": lon,
  "start_date": date,
  "end_date": date,
  "daily": ["rain_sum", "precipitation_hours", "temperature_2m_max", "temperature_2m_min", "precipitation_sum", "weather_code"]
  }
  responses = openmeteo.weather_api(url, params=params)
  response = responses[0]

  daily = response.Daily()
  daily_rain_sum = daily.Variables(0).ValuesAsNumpy()
  daily_precipitation_hours = daily.Variables(1).ValuesAsNumpy()
  daily_temperature_2m_max = daily.Variables(2).ValuesAsNumpy()
  daily_temperature_2m_min = daily.Variables(3).ValuesAsNumpy()
  daily_precipitation_sum = daily.Variables(4).ValuesAsNumpy()
  daily_weather_code = daily.Variables(5).ValuesAsNumpy()

  return [daily_rain_sum, daily_precipitation_hours, daily_temperature_2m_max, daily_temperature_2m_min,
          daily_precipitation_sum, daily_weather_code]
```

**Переведем столбец 'Order Date' в формат, необходимый для запроса и вызовем функцию get_weather_with_coord1 для датафрейма**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
df['Order Date'] = df['Order Date'].dt.strftime('%Y-%m-%d')

weather_data = df.apply(lambda row: get_weather_with_coord1(row['Order Date'], row['Latitude'], row['Longitude']), axis=1)

df["daily_rain_sum"] = weather_data.apply(lambda x: x[0][0])
df["daily_precipitation_hours"] = weather_data.apply(lambda x: x[1][0])
df["daily_precipitation_sum"] = weather_data.apply(lambda x: x[4][0])
df["daily_weather_code"] = weather_data.apply(lambda x: x[5][0])
df['daily_temperature_2m_max'] = weather_data.apply(lambda x: x[2][0])
df['daily_temperature_2m_min'] = weather_data.apply(lambda x: x[3][0])

df
```

**Столбец 'daily_rain_sum' отвечает за количество мм осадков (конкретно дождя). Следовательно, иожно добавить новый бинарный признак 'Rain', в котором будет 1, если дождь был и 0 - если нет**

```python
df['Rain'] = (df['daily_rain_sum'] > 0).astype(int)
```
